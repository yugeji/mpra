```python
%matplotlib inline
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data_f = open("chrm1_data.txt").readlines()

locations = [l.split("-")[0][5:] for l in data_f]
sharpr_score = sorted([float(l.split("\t")[1]) for l in data_f])
```

```python
plt.plot(locations, sharpr_score)
```

```python
#Normal distribution
plt.rcParams["figure.figsize"] = (20,5)

pdf = stats.norm.pdf(sharpr_score, np.mean(sharpr_score), np.std(sharpr_score))
plt.plot(sharpr_score, pdf, '-o')
```
